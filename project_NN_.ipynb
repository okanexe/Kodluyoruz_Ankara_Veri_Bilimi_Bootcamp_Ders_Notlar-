```python
data = pd.read_csv("data_last_upgrade.csv")
data.head(0)
```

```python
# Dates columunu Ay yıl gün olarak ayırma

import datetime as dt

Dates_Converted = []
Hafta_Gunu = []

for i in data["Dates"]:
    
    # print(i) # 2015-07-22
    yil = i[0:4]
    ay = i[5:7]
    gun = i[8:10]    
    
    # string gun-ay-yıl bılgısını date formatına cevirme
    d = dt.datetime.strptime("{}-{}-{}".format(gun,ay,yil), "%d-%m-%Y")
    
    # Convert datetime object to date object.
    d = d.date()
    Dates_Converted.append(d) #listeye ekleme
```

```python
# 0 Pazartesi 
# 1 Salı 
# 2 Çarşamba
# 3 Perşembe
# 4 Cuma
# 5 Cumartesi
# 6 Pazar

for i in Dates_Converted:
    
    m= i.weekday()
    
    if m>=0 and m<5:
        Hafta_Gunu.append("Hafta İçi")
    else:
        Hafta_Gunu.append("Hafta Sonu") 

Hafta_Gunu   
```
