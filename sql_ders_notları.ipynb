```python
import sqlite3 as sql
```

```python
kitaplarvt = sql.connect('kitaplar.sqlite')
```

```python
im = kitaplarvt.cursor()
```

```python
im.execute("SELECT KitapAdi FROM kitaplar")
im.fetchall()
```

```python
im.execute("SELECT Fiyati FROM kitaplar")
im.fetchall()
```

```python
im.execute("SELECT * FROM kitaplar")
im.fetchall()
```

```python
im.execute("SELECT * FROM kitaplar WHERE Fiyati < 30")
im.fetchall()
```

```python
im.execute("SELECT * FROM kitaplar WHERE Yazar LIKE 'A_%' and Fiyati < 30")
im.fetchall()
```

```python
im.execute("SELECT KitapAdi, Fiyati FROM kitaplar WHERE Fiyati > (SELECT min(Fiyati) FROM kitaplar WHERE Yazar LIKE 'O_%')")
im.fetchall()
```
